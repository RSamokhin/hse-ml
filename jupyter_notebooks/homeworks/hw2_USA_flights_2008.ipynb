**5. Определите, как распределено количество рейсов в зависимости от времени дня.**

**Выберите все верные утверждения.**
 - Распределение полетов по часам (0-23) распределено нормально
 - Распределение полетов по часам (0-23) распределено равномерно
 - В период с 0 am до 4 am полетов на порядок меньше, чем в 7 pm
 - В 5 утра перелетов больше, чем в 10 вечера


```python
# Ваш код здесь

def getHour(d):
    return int(d/100)

flight_nn = flights_df[flights_df.DepTime >= 0]
flight_nn['HOUR'] = flight_nn['DepTime'].apply(getHour)
```

```python

x = np.sort(flight_nn.HOUR.unique())
y = flight_nn.groupby('HOUR').FlightNum.count().values
print(flight_nn.groupby('HOUR').FlightNum.count())


plt.plot(x, y, color = 'red', linestyle = 'dashed', marker = 'o',
         markeredgecolor = 'black', markersize = 10, 
         linewidth = 3, label = r"$y = x^2$")
plt.xlabel(r"$hour$")
plt.ylabel(r"$flights count$")
plt.title('Flights Count by Hour')
plt.show()
flight_nn.groupby('HOUR')['FlightNum'].agg('count').plot()
flight_nn.groupby('HOUR')['FlightNum'].agg('count')
```
